# Logistic Regression

This article gives a basic overview of one of the most popular machine learning algorithms. Logistic Regression is a supervised classification algorithm.

It might be confusing at first as to why is this named as a regression algorithm even though this is a classification algorithm, but the algorithm actually outputs the probability of occurence of an event which is a continous variable (which is later converted to either y=0 or y=1 based on a threshold value). This is why 'regression' appears in the name.

Examples: email spam filtering, image classification, bank fraud detection, etc.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

Before jumping into the intricacies of Logistic Regression, let's just try fitting a linear model to our classification data and see how the model performs.

Consider a simple dataset where the features would be salaries of the loan applicants and y would represent whether the loan is approved or not. This would be a binary classification with the two classes being y = 0 ( loan not approved ) and y =1 ( loan approved ).

```python
a = (np.array(np.arange(1,21)*1000).tolist())
b = np.array(np.zeros(int(len(a)/2)).tolist() + np.ones(int(len(a)/2)).tolist())

fig, ax=plt.subplots(figsize=(9,5))
plt.scatter(a,b)
plt.ylim(-0.05,1.05)
plt.xlabel('Salary')
plt.ylabel('Loan Approval')
plt.show()
```

It's easy to deduce from the plot that only those applicants whose salary is greater than 10,000 would have their applications approved. If we fit a linear model to this plot using Ordinary Least Squares method, it'd look like this:

```python
a = (np.array(np.arange(1,21)*1000).tolist())
b = np.array(np.zeros(int(len(a)/2)).tolist() + np.ones(int(len(a)/2)).tolist())

fig, ax=plt.subplots(figsize=(9,5))
ax.scatter(a,b,alpha= 0.3)
sns.regplot(a,b,ci=0,ax=ax, label='Linear Model')
plt.plot(np.linspace(0,20000,20000), np.zeros(20000)+0.5,'--k', alpha=0.5)
plt.plot(np.linspace(10500,10500,50), np.linspace(0,1.6,50), '--k', alpha=0.5)
# plt.xlim(0,30)
plt.ylim(-0.05,1.05)
plt.xlabel('Salary')
plt.ylabel('Loan Approval')
plt.legend()
plt.show()
```

The equation of the linear model is:
### $g({\theta}) = {\theta}_{0} + {\theta}_{1}{x}_{1}$

- where ${\theta}_{0}$ and ${\theta}_{0}$ are parameters and $x_1$ represents salary

If we set a threshold value of y=0.5, as seen in the graph, then all the salary values that are greater than 10,000 would be approved for loan according to this linear model. This model actually does a good job of capturing the relationship between the feature and the label. But there's a catch, what if there are outliers in our data?

Let's now plot the same data along with outliers and see how the linear model performs.

```python
a = np.array(np.arange(1,21).tolist() + [42,48])
b = np.array(np.zeros(int(len(a)/2)).tolist() + np.ones(int(len(a)/2)).tolist())

fig, ax=plt.subplots(figsize=(9,5))
ax.scatter(a,b,alpha= 0.3)
sns.regplot(a,b,ci=0,ax=ax)
plt.plot(np.linspace(0,50,50), np.zeros(50)+0.5,'--k')
plt.plot(np.linspace(13.8,13.8,50), np.linspace(0,1.6,50), '--k')

plt.ylim(-0.05,1.05)
plt.show()
```

The model now tilts a little inorder to accomodate the outliers. This shifts the threshold salary from 10,000 to around 13,000. As a result of this, the new linear model would misclassify all the salaries between 10,000 and 13,000  as y=0 while in fact they should've been classified as y=1.

This is when the linear model fails to deliver. Another drawback of using linear models in classification problems is that the linear model does not always return values between 0 and 1, and any value beyond the range of [0,1] would make no sense because the value of probability always lies between 0 and 1.

Inorder to solve this, we'd need a function which takes this linear equation score and outputs a number between 0 and 1 and this number would represent the probability that the observation belongs to class y=1. This function is called sigmoid function or logistic function. The formula and graph would be as follows:

$$h(z) = \frac{1}{1+e^{-z}}$$


```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
fig, ax=plt.subplots(figsize=(9,5))
plt.title('Sigmoid Function')
plt.plot(x, z)
plt.xlabel("z")
plt.ylabel("Sigmoid(z)")
plt.grid()
  
plt.show()
```

It is worth noting from the graph that the sigmoid function :
- is almost linear between (-3,3)
- is 0.5 when z = 0
- asymptotically reaches 0 when z < -3
- asymptotically reaches 1 when z > 3
<br>
<br>

If $\theta$ represents a matrix of parameters $\theta_1$, $\theta_2$... $\theta_n$ and X represents all the features $x_1$, $x_2$... $x_n$, then logistic regression can be expressed in the form of a matrix as follows:

$$h(\theta^T  X) = \frac{1}{1 + e^{-\theta^T X}}$$
<br>
To reiterate, $h(\theta^T  X)$ outputs a number between 0 and 1 which gives us the probability of X belonging to class y = 1. We can set a thershold of y = 0.5, then:
- $y = 1$, when $h(\theta^T  X )\ge$  0.5 $\implies \theta^T  X \ge 0$
- $y = 0$, when $h(\theta^T  X )\lt$  0.5 $\implies \theta^T  X \lt 0$


To get a better understanding let's use a built-in dataset called 'penguins' from Seaborn, it has data about bill features of penguins along with the name of the species. It'd be interesting to see if we can segregate 'Adelie' and 'Gentoo' species by using just the bill depth and bill length of the penguins using logistic regression.


Bill length and bill depth would be the two features and the species would be the label. This is how the scatter plot would look like:

```python

data = sns.load_dataset('penguins')
data = data[data['species'].isin(['Adelie', 'Gentoo'])]
data.dropna(inplace=True)

plt.figure(figsize=(9,5))
plt.title('Penguins: bill_depth vs bill_length')
plt.xlabel('Bill length in mm')
plt.ylabel('Bill depth in mm')
sns.scatterplot(data=data, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.show()
```

The two classes can simply be segragated by a straight line as seen in the graph, and the equation of this straight line is given by:

$g({\theta}) = {\theta}_{0} + {\theta}_{1}{x}_{1} + {\theta}_{2}{x}_{2}$

- x1 represents bill length
- x2 represents bill depth
- ${\theta}_{0}, {\theta}_{1}, {\theta}_{2}$ represent the three parameters

This line that separates the two classes is called the 'Decision Boundary'. 

```python
plt.figure(figsize=(9,5))
plt.title('Decision Boundary')
plt.xlabel('Bill length in mm')
plt.ylabel('Bill depth in mm')
sns.scatterplot(data=data, x='bill_length_mm', y='bill_depth_mm', hue='species')
x = np.linspace(36,55)
sns.lineplot(x, x/2 - 4.5, color = 'purple', label='decision boundary')
plt.show()
```

The decision boundary is a property of the weights and not a property of the dataset. The parameter values that define the decision boundary can be easily determined by fitting a logistic regression model to the data and then extracting the coefficients and intercept.

```python
X = data[['bill_length_mm', 'bill_depth_mm']]
y = data['species']
model = LogisticRegression()
model.fit(X,y)
print(f'\u03B8\u2080 = {model.intercept_[0]}')
print(f'\u03B8\u2081 = {model.coef_[0][0]}')
print(f'\u03B8\u2082 = {model.coef_[0][1]}')
```

For simplicity, lets consider 
- ${\theta}_{0}$ = -16 
- ${\theta}_{1}$ = 1 
- ${\theta}_{2}$ = -2

The linear equation would look like this after plugging in the parameter values:
$g({\theta}) = -16 + {x}_{1} - 2{x}_{2}$

Given any new point $({x}_{1}, {x}_{2})$, we can simply plot the point on the graph and the decision boundary would tell us which class does the point belong to, it's that simple.

## Let's discuss decision boundary in more detail:
A logistic regression model gives us the probability of a point belonging to a certain class. In our example of penguins, when $g({\theta})$ is passed through the sigmoid function, it returns us the probability that the point belongs to Gentoo species, p. The probability that this point belongs to class 'Adelie' would simply be 1-p.

Logistic Regression provides us with a probabilty space which can be divided into three parts based on the decision boundary:
- All the points that lie towards the right of the decision boundary would have $g({\theta}) \gt$ 0 and hence $h( g({\theta}) ) \gt$ 0.5. Meaning, the points that lie on the right of the decision boundary have more porbabilty of belonging to the 'Gentoo' category.
- Similarly, all the points that lie towards the left of the decision boundary would have $g({\theta}) \lt$ 0 and hence $h( g({\theta}) ) \lt$ 0.5. Meaning, the points that lie on the left of the decision boundary have more porbabilty of belonging to the 'Adelie' category.
- All the points that lie on the decision boundary would have equal probability of belonging to each of the classes. But since we have defined the threshold as $y = 1$, when $h(\theta^T  X )\ge$  0.5, then all the points on the decision boundary would belong to class Gentoo.

This was just a simple example where there were just two categories and they could be separated with just a straight line, but in real life we'd come across much more complex problems. If the two classes required a more complex decision boundary, say a circle, then instead of just using a linear equation we'd use polynomial regression to increase the complexity.

## Cost Function:

A cost function gives an estimate of how badly a model is performing. But before we define the cost function for logistic regression, let's have a look at two graphs that give us an idea about what we are looking for.

```python
x = np.linspace(0,1,10000)
y = -np.log(x)
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.lineplot(x,y)
plt.ylim(0,9)
plt.title('Cost function when y = 1')
plt.xlabel(f'$h_\u03B8$(x)')
plt.ylabel(f'cost (h(\u03B8), y) ')
plt.grid()
plt.tight_layout()


y = -np.log(1-x)
plt.subplot(1,2,2)
sns.lineplot(x,y)
plt.ylim(0,9)
plt.title('Cost function when y = 0')
plt.xlabel(f'$h_\u03B8$(x)')
plt.ylabel(f'cost (h(\u03B8), y) ')
plt.grid()
plt.tight_layout()
```

If one can stare at these graphs for long enough, then it's pretty easy to understand what they signify. 

When the actual label is y = 1, but our model predicts otherwise and says $\hat{y} = 0$ then our model is misclassifying the point and hence there is a huge error (or cost). This is what is happening in the first plot. The cost tends to infininty when $\hat{y} = 0$ while in reality y = 1, and the cost keeps on decreasing when our model starts producing probability values closer to 1.


$$cost( h_\theta(x), y=1) = - log(h_\theta(x))$$ 
$$cost( h_\theta(x), y=0) = - log(1 - h_\theta(x))$$  

Simplified cost function can be written as:
$$cost( h_\theta(x), y) = - y log(h_\theta(x)) - (1-y) log(1-h_\theta(x))$$ 

This simplified cost function gives us the error at a single point. The average of this error across all the points gives us the total error of the model and it's called Cross Entropy Loss.

## Conclusion

Logistic Regression is regression algorithm used when the label is a categorical variable. A logistic regression model consists of two parts, namely, a linear equation and an activation function (sigmoid function). The more complex the decision boundary is, the more complex our linear equation would be. The score produced from the linear equation is passed through a sigmoid function which gives us the probability value of the point belonging to class y = 1. The parameters that define the model are optimized through a technique called Gradient Descent and the performance of the model is computed through the cross entropy loss function.
